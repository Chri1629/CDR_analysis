## Merge dei due dataset
elinimando gli studenti duplicati

```python
import pandas as pd
```

```python
df1 = pd.read_csv("student-mat.csv")
df1.head()
```

```python
df2 = pd.read_csv("student-por.csv")
df2.head()
```

```python
df = pd.concat([df1,df2], ignore_index=True)
df.head()
```

```python
c1 = df1['school'].count()
c1
```

```python
c2 = df2['school'].count()
c2
```

```python
#somma degli studenti
c1 + c2
#sudenti ridondanti 382 
#==> 1044 - 382 = 662
```

```python
df.drop_duplicates(subset=["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"], inplace=True)
df['school'].count()
```
