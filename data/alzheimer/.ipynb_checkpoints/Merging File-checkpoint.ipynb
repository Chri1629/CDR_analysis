## Merge dei due dataset
elinimando gli studenti duplicati

```python
import pandas as pd
```

```python
df1 = pd.read_csv("oasis_cross-sectional.csv")
df1.head()
```

```python
df2 = pd.read_csv("oasis_longitudinal.csv")
df2.head()
```

```python
#elimino le colonne inutili/ridondanti
df21 = df2.drop(columns=['Subject ID', 'Group', 'Visit','MR Delay'])
df21.head()
```

```python
# elimino tutti i valori null
df21 = df21.dropna()
df21.head()
```

```python
df11 = df1.drop(columns=['Delay'])
df11.rename(columns={'Educ' : 'EDUC', 'ID' : 'MRI ID'}, inplace=True)
df11.head()
```

```python
# elimino le righe null
df11 = df11.dropna()
df11.head()
```

```python
df = pd.concat([df11,df21], ignore_index=True)
df.head()
```

```python
c1 = df11['MRI ID'].count()
c1
```

```python
c2 = df2['MRI ID'].count()
c2
```

```python
#somma delle righe
c1 + c2
# 436 + 373 - 220 (record null) = 589
```

```python
df.info()
```

```python
df21.isnull().sum()
```

```python
df.to_csv("oasis_merge.csv")
```
